```python
data3 = data[['Gender', 'Customer Type', 'Type of Travel',
'Class', 'satisfaction']]
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
data3.loc[:,:] = data3.loc[:,:].apply(label_encoder.fit_transform)
```

```python
data3.head()
```

```python
data[list(data3.columns)] = data3
data = data.apply(pd.to_numeric, errors='coerce')
data.head()
```

```python
corr = pd.DataFrame(data.corr()['satisfaction']).drop('satisfaction', axis=0)
corr.plot(kind='barh',grid=True,figsize=(15,15))
```

```python
feature_columns = []
for x in corr.index:
    if corr.loc[x].values < -0.4:
        feature_columns.append(x)
    elif corr.loc[x].values > 0.2:
        feature_columns.append(x)
```

```python
print(feature_columns)
```

```python
X = data[feature_columns]
y = data['satisfaction']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
X.shape
```

```python
y.shape
```

```python
x_train = X.iloc[:72733]
```

```python
y_train = y.iloc[:72733]
```

```python
x_test = X.iloc[72733:103905]
y_test = y.iloc[72733:103905]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
```

```python
print("Training Accuracy :", lr.score(x_train, y_train))
print("Testing Accuracy :", lr.score(x_test, y_test))
```

```python
print('Logistic Regression Accuracy Score:',
accuracy_score(lr_pred,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
dt_pred = dt.predict(x_test)
```

```python
print("Training Accuracy :", dt.score(x_train, y_train))
print("Testing Accuracy :", dt.score(x_test, y_test))
```

```python
print('Decision Tree Accuracy Score:',
accuracy_score(dt_pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
rf_pred = rf.predict(x_test)
```

```python
print("Training Accuracy :", rf.score(x_train, y_train))
print("Testing Accuracy :", rf.score(x_test, y_test))
```

```python
print('Random Forest Accuracy Score :', rf.score(x_test,y_test))
```
